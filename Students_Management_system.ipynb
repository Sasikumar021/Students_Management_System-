```python
import mysql.connector

# Database Connection
mydb = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Sasi@021##',
    database='student_management'
)

mycursor = mydb.cursor()

# Add Student
def add_student():
    student_id = int(input("Enter Student ID: "))
    name = input("Enter Student Name: ")
    age = int(input("Enter Age: "))
    gender = input("Enter Gender: ")
    course = input("Enter Course: ")
    mobile = input("Enter Mobile Number: ")
    marks = float(input("Enter Marks: "))

    query = "INSERT INTO students VALUES(%s,%s,%s,%s,%s,%s,%s)"
    values = (student_id, name, age, gender, course, mobile, marks)

    mycursor.execute(query, values)
    mydb.commit()

    print("Student Added Successfully!")

def update_marks():
    student_id = int(input("Enter Student ID: "))
    marks = float(input("Enter New Marks: "))

    query = "UPDATE students SET marks=%s WHERE student_id=%s"

    mycursor.execute(query, (marks, student_id))
    mydb.commit()

    if mycursor.rowcount > 0:
        print("Marks Updated Successfully")
    else:
        print("Student Not Found")

def search_student():
    student_id = int(input("Enter Student ID: "))

    query = "SELECT * FROM students WHERE student_id=%s"

    mycursor.execute(query, (student_id,))

    result = mycursor.fetchone()

    if result:
        print("\n========= Student Details =========")
        print("Student ID :", result[0])
        print("Name       :", result[1])
        print("Age        :", result[2])
        print("Gender     :", result[3])
        print("Course     :", result[4])
        print("Mobile     :", result[5])
        print("Marks      :", result[6])
    else:
        print("Student Not Found")

def display_students():

    query = "SELECT * FROM students"

    mycursor.execute(query)

    result = mycursor.fetchall()

    print("\n=========== Student Records ===========")

    for row in result:
        print("----------------------------------")
        print("Student ID :", row[0])
        print("Name       :", row[1])
        print("Age        :", row[2])
        print("Gender     :", row[3])
        print("Course     :", row[4])
        print("Mobile     :", row[5])
        print("Marks      :", row[6])

def delete_student():

    student_id = int(input("Enter Student ID: "))

    query = "DELETE FROM students WHERE student_id=%s"

    mycursor.execute(query, (student_id,))
    mydb.commit()

    if mycursor.rowcount > 0:
        print("Student Deleted Successfully")
    else:
        print("Student Not Found")

while True:

    print("\n========== STUDENT MANAGEMENT SYSTEM ==========")
    print("[1] Add Student")
    print("[2] Update Marks")
    print("[3] Search Student")
    print("[4] Display All Students")
    print("[5] Delete Student")
    print("[6] Exit")

    choice = int(input("\nEnter Your Choice: "))

    if choice == 1:
        add_student()

    elif choice == 2:
        update_marks()

    elif choice == 3:
        search_student()

    elif choice == 4:
        display_students()

    elif choice == 5:
        delete_student()

    elif choice == 6:
        print("Thank You")
        mycursor.close()
        mydb.close()
        break

    else:
        print("Invalid Choice")
```
